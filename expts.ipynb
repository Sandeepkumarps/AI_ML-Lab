# Matrix Addition


```python
r1=int(input("Enter the row of first matrix:"))
c1=int(input("Enter the column of first matrix:"))

print("Enter elements to the first matrix:")
A=[]
for i in range(r1):
    row=[]
    for j in range(c1):
        row.append(int(input()))
    A.append(row)

print("Enter elements to the second matrix:")
B=[]
for i in range(r1):
    row=[]
    for j in range(c1):
        row.append(int(input()))
    B.append(row)
    
print("Matrix A:")
for i in range(r1):
    for j in range(c1):
        print(A[i][j], end=" ")
    print()
    
print("Matrix B:")
for i in range(r1):
    for j in range(c1):
        print(B[i][j], end=" ")
    print()
    
result=[]
for i in range(r1):
    row=[]
    for j in range(c1):
        row.append(0)
    result.append(row)

for i in range(r1):
    for j in range(c1): 
        result[i][j]=A[i][j]+B[i][j]
    
print("Result Matrix:")
for i in range(r1):
    for j in range(c1): 
        print(result[i][j] , end=" ")
    print()
```

# MEAN

```python
def mean(L):
    sum=0
    n=len(L)
    for ele in L:
        sum+=ele
    return sum/n
    
l=[int(x) for x in input("Enter ehe elements:").split()]
print("Mean:",mean(l)) 
```

# MEDIAN

```python
def Median(n,L):
    temp=0
    for i in range(0,n):
        for j in range(i+1,n):
            if(L[i]>L[j]):
                temp=L[i]
                L[i]=L[j]
                L[j]=temp
    
    if(n%2!=0):
        return(L[n//2])
        
    else:
        return((L[(n)//2]+L[(n+1)//2])/2)
        
    
n=int(input("Enter the length of List:"))
print("Enter the elements to the List:")

L=[int(input())for i in range(n)]
sortList=Median(n,L)

print("Median=",sortList)
```

# MODE


```python
def Mode(L):
    result=[]
    dict1=dict.fromkeys(L,0)
    for i in L:
        if i in dict1:
            dict1[i]+=1
            
    for key in dict1:
        if dict1[key]==max(dict1.values()):
            result.append(key)
    return result
    
n=int(input("Enter the lenght of List:"))
L=[int(input()) for i in range(n)]
print("Mode=",Mode(L))   
```

# VARIANCE & STANDARD DEVIATION

```python
def variance(l):
    xbar=mean(l)
    newl=[(x-xbar)**2 for x in l]
    return sum(newl)/(len(l)-1)
    
l=[int(x) for x in input("Enter ehe elements:").split()]
print("variance=",variance(l)) 
print("Standard variance:",variance(l)**(1/2))
```

# Password Checking

```python
import re
def checkPass(password):
    errors=[]
    if(len(password)<=8):
        errors.append("At least 8 elements")
    if not re.search('[A-Z]',password):
        errors.append('at least one uppercase')
    if not re.search('[a-z]',password):
        errors.append('at least one lowercase')
    if not re.search('[0-9]',password):
        errors.append('at least one digit')
    if not re.search('[!@#$%^&*]',password):
        errors.append('at least one specialCharacters')
    
    if errors:
        return ("Try again".join(errors))
    else:
        return ("Valid")

password=input("Enter Password")
checkPass(password)
```

```python
def timeOfDay(n):
    if n in range(1,4):
        return "Late Night"
    elif n in range(4,7):
        return "Early Morning"
    elif n in range(7,12):
        return "Morning"
    elif n in range(12,15):
        return "After Noon"
    elif n in range(15,18):
        return "Evening"
    elif n in range(18,25) or n==0:
        return "Night"
```

```python
df['timeofday']=df['hour'].apply(lambda x:timeOfDay(x))
```

```python
def monthToSeasons(n):
    if n in [9,10,11]:
        return "Spring"
    elif n in [12,1,2]:
        return "Summer"
    elif n in [3,4,5]:
        return "Autumn"
    elif n in [6,7,8]:
        return "Winter"
```

```python
df['seasons']=df['month'].apply(lambda x:monthToSeasons(x))
```
